```python
df = df_ords_prods_mergd[:1000000]
```

```python
# creating a function to categorize items in to price ranges low, medium, high, and not enough data
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Creating a new list to correspond with this request
result2 = []

for value in df_ords_prods_mergd["order_dow"]:
  if value == 0 and 1:
    result2.append("Busiest days")
  elif value == 3 and 4:
    result2.append("Slowest days")
  else:
    result2.append("Regularly busy")
```
