# 04. Merging Dataframes

```python
# Importing clean customers dataframe 
path = r'C:\Users\simpe\Instacart Basket Analysis'
customers_clean = pd.read_pickle(os.path.join(path,'02 Data','Prepared Data', 'customers_clean.pkl'))
```

```python
customers_clean.shape
```

```python
# Importing ords_prods dataframe

ords_prods = pd.read_pickle(os.path.join(path,'02 Data','Prepared Data', 'ords_prods_merge_4.8.pkl'))
```

```python
# Converting user_id datatype in ords_prods dataframe 

ords_prods['user_id'] = ords_prods['user_id'].astype('str')
```

```python
# Merging customer and ords_prods dataframes 

ords_prods_cust = ords_prods.merge(customers_clean, on = 'user_id')
```

```python
ords_prods_cust.shape
```

```python
pd.options.display.max_columns = None
ords_prods_cust.head()
```

```python
#changing data types of integer columns to save space and memory
ords_prods_cust['aisle_id'] = ords_prods_cust['aisle_id'].astype('int8')
ords_prods_cust['department_id'] = ords_prods_cust['department_id'].astype('int8')
ords_prods_cust['prices'] = ords_prods_cust['prices'].astype('float16')
ords_prods_cust['age'] = ords_prods_cust['age'].astype('int8')
ords_prods_cust['number_of_dependents'] = ords_prods_cust['number_of_dependents'].astype('int8')
ords_prods_cust['income'] = ords_prods_cust['income'].astype('int32')
```

```python
ords_prods_cust.head()
```

```python
# Exporting combined dataframe. 
ords_prods_cust.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_customers_merged.pkl'))
```
